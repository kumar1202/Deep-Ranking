```python
from PIL import Image
import requests
from io import BytesIO
```

```python
file=open('query_and_triplets.txt')
x=file.read()
print(x)
```

```python
import re
#dir(re)
li1=re.findall('http://.*',x)
li1
```

```python
pixlist=[]
print(len(li1))
```

```python
for link in li1[:15]:
    response=requests.get(link)
    img = Image.open(BytesIO(response.content))
    pix_val = list(img.getdata())
    pixlist.append(pix_val)
```
